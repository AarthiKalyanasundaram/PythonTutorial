```python
plt.title("Graph between Graduates and Income")
plt.xlabel("Number of Graduates")
plt.ylabel("Income")
plt.scatter(dataframe.graduates, dataframe.income)
plt.show()
```

```python
plt.title("Graph between Graduates and Income")
plt.xlabel("Graduates")
plt.ylabel("Income")
plt.plot(dataframe.graduates.sort_values().head(10), dataframe.income.sort_values().head(10))
plt.show()
```

```python
#To increase the size of the plot, 

plt.figure(figsize = (10,8))
plt.title("Graph between Graduates and Income")
plt.xlabel("Graduates")
plt.ylabel("Income")
plt.plot(dataframe.graduates.sort_values().head(10), dataframe.income.sort_values().head(10))
plt.show()
```

```python
#Formatting and Styling

plt.figure(figsize = (10,8))
plt.title("Graph between Graduates and Income")
plt.xlabel("Graduates")
plt.ylabel("Income")
plt.plot(dataframe.graduates.sort_values().head(10), dataframe.income.sort_values().head(10), "g^")
plt.show()
```

```python
#plotting 2 consecutive plots

plt.figure(figsize = (10,8))
plt.title("Graph between Graduates and Income")
plt.xlabel("Graduates")
plt.ylabel("Income")
plt.plot(dataframe.graduates.sort_values().head(10), dataframe.percent_senior.sort_values().head(10), "g^")
plt.show()
plt.plot(dataframe.crime_rate.sort_values().head(10), dataframe.percent_senior.sort_values().head(10), "bx--")
plt.show()
```

```python
#Adding a Legend

plt.figure(figsize = (10,8))
plt.title("Graph")
plt.xlabel("Graduates")
plt.ylabel("Income")
plt.plot(dataframe.graduates.sort_values().head(10), dataframe.percent_senior.sort_values().head(10), "g^", label = "Graduate Vs Income")
plt.plot(dataframe.crime_rate.sort_values().head(10), dataframe.percent_senior.sort_values().head(10), "bx--", label = "Graduate Vs Income")

plt.legend()
plt.show()
```

```python
#Adding a Legend

plt.figure(figsize = (10,8))
plt.title("Graph")
plt.xlabel("Percent Senior")
plt.ylabel("Income")
plt.plot(dataframe.percent_senior.sort_values().head(10), dataframe.graduates.sort_values().head(10), "g^", label = "Percent senior Vs Graduate")
plt.plot(dataframe.percent_senior.sort_values().head(10), dataframe.crime_rate.sort_values().head(10), "bx--", label = "Percent senior Vs Crime rate")

plt.legend()
plt.show()
```

```python
#Subplot

plt.subplot(1,2,1)  # row, column, index
plt.plot(dataframe.work_force, dataframe.income, "go")
plt.title("Income vs Work Force")

## plt.subplot(1,2,2) # row, column, index
plt.subplot(1,2,2).label_outer()

plt.plot(dataframe.hospital_beds, dataframe.income, "r^")
plt.title("Income vs Hospital Beds")

plt.suptitle("Sub Plots") # Add a centered title to the figure.
plt.show()
```

```python
#Plotting a Subplot

plt.subplot(3,1,1)  # row, column, index
plt.plot(dataframe.work_force, dataframe.income, "go")
plt.title("Income vs Work Force")

## plt.subplot(1,2,2) # row, column, index
plt.subplot(3,1,2).label_outer()

plt.plot(dataframe.hospital_beds, dataframe.income, "r^")
plt.title("Income vs Hospital Beds")

plt.subplot(3,1,3).label_outer()


plt.plot(dataframe.percent_senior, dataframe.income, "b*")
plt.title("Income vs Percent Senior")


plt.suptitle("Sub Plots") # Add a centered title to the figure.
plt.show()
```

```python
#Plotting a Subplot

plt.subplot(1,3,1)  # row, column, index
plt.plot(dataframe.work_force, dataframe.income, "go")
plt.title("Income vs Work Force")

## plt.subplot(1,2,2) # row, column, index
plt.subplot(1,3,2).label_outer()

plt.plot(dataframe.hospital_beds, dataframe.income, "r^")
plt.title("Income vs Hospital Beds")

plt.subplot(1,3,3).label_outer()


plt.plot(dataframe.percent_senior, dataframe.income, "b*")
plt.title("Income vs Percent Senior")


plt.suptitle("Sub Plots") # Add a centered title to the figure.
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6)) #creating a grid of 2 rows, 2 columns and 6x6 figure size
ax[0,0].plot(dataframe.work_force, dataframe.income, "go") # The top-left axes
ax[0,1].plot(dataframe.work_force, dataframe.income, "bo") # The top-right axes
ax[1,0].plot(dataframe.work_force, dataframe.income, "yo") # The bottom-left axes
ax[1,1].plot(dataframe.work_force, dataframe.income, "ro") # The bottom-right axes

plt.show()
```
